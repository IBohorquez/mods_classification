```python
N_iters = df_crossval_results['N_iteration'].unique().shape[0]

train_acc_avg_iters = []
train_prec_avg_iters = []
train_sens_avg_iters = []
train_spec_avg_iters = []
train_f1_avg_iters = []
train_roc_auc_avg_iters = []
train_qwk_avg_iters = []

val_acc_avg_iters = []
val_prec_avg_iters = []
val_sens_avg_iters = []
val_spec_avg_iters = []
val_f1_avg_iters = []
val_roc_auc_avg_iters = []
val_qwk_avg_iters = []


test_acc_avg_iters = []
test_prec_avg_iters = []
test_sens_avg_iters = []
test_spec_avg_iters = []
test_f1_avg_iters = []
test_roc_auc_avg_iters = []
test_qwk_avg_iters = []

for i in range(N_iters):
    

    df_temp = df_crossval_results[df_crossval_results['N_iteration']==i+1]

    train_acc_avg_fold = []
    train_prec_avg_fold = []
    train_sens_avg_fold = []
    train_spec_avg_fold = []
    train_f1_avg_fold = []
    train_roc_auc_avg_fold = []
    train_qwk_avg_fold = []

    val_acc_avg_fold = []
    val_prec_avg_fold = []
    val_sens_avg_fold = []
    val_spec_avg_fold = []
    val_f1_avg_fold = []
    val_roc_auc_avg_fold = []
    val_qwk_avg_fold = []


    test_acc_avg_fold = []
    test_prec_avg_fold = []
    test_sens_avg_fold = []
    test_spec_avg_fold = []
    test_f1_avg_fold = []
    test_roc_auc_avg_fold = []
    test_qwk_avg_fold = []


    for k in range(4):
        for m in ['train', 'val', 'test']:
            eval(m + '_' + "acc_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'Accuracy'])")
            eval(m + '_' + "prec_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'Precision'])")
            eval(m + '_' + "sens_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'Sensitivity'])")
            eval(m + '_' + "spec_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'Specificity'])")
            eval(m + '_' + "f1_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'F1_score'])")
            eval(m + '_' + "roc_auc_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'ROC AUC Score'])")
            eval(m + '_' + "qwk_avg_fold.append(df_temp[df_temp['k_fold']==k+1].iloc[-1][m + ' ' + 'Quadratic Weighted Kappa'])")

    train_acc_avg_iters.append(mean(train_acc_avg_fold))
    train_prec_avg_iters.append(mean(train_prec_avg_fold))
    train_sens_avg_iters.append(mean(train_sens_avg_fold))
    train_spec_avg_iters.append(mean(train_spec_avg_fold))
    train_f1_avg_iters.append(mean(train_f1_avg_fold))
    train_roc_auc_avg_iters.append(mean(train_roc_auc_avg_fold))
    train_qwk_avg_iters.append(mean(train_qwk_avg_fold))
    
    val_acc_avg_iters.append(mean(val_acc_avg_fold))
    val_prec_avg_iters.append(mean(val_prec_avg_fold))
    val_sens_avg_iters.append(mean(val_sens_avg_fold))
    val_spec_avg_iters.append(mean(val_spec_avg_fold))
    val_f1_avg_iters.append(mean(val_f1_avg_fold))
    val_roc_auc_avg_iters.append(mean(val_roc_auc_avg_fold))
    val_qwk_avg_iters.append(mean(val_qwk_avg_fold))

    test_acc_avg_iters.append(mean(test_acc_avg_fold))
    test_prec_avg_iters.append(mean(test_prec_avg_fold))
    test_sens_avg_iters.append(mean(test_sens_avg_fold))
    test_spec_avg_iters.append(mean(test_spec_avg_fold))
    test_f1_avg_iters.append(mean(test_f1_avg_fold))
    test_roc_auc_avg_iters.append(mean(test_roc_auc_avg_fold))
    test_qwk_avg_iters.append(mean(test_qwk_avg_fold))
```
